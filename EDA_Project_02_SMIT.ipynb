# **The dataset contains no missing values. We will now proceed to explore the data, detect outliers, and uncover key insights.**

```python
data.head()
```

```python
# Ensure column is string
data["Streams"] = data["Streams"].astype(str)

# Remove commas and convert to integer
data["Streams"] = data["Streams"].str.replace(",", "").astype(int)

# Check
data["Streams"].head()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data["Streams"], bins=30, kde=True)  # bins define frequency groups
plt.title("Distribution of Streams")
plt.xlabel("Streams")
plt.ylabel("Frequency")
plt.show()
```

```python
# Checking outliers:

plt.figure(figsize=(10,4))
sns.boxplot(x=data["Streams"])
plt.title("Boxplot of Streams")
plt.show()
```

```python
data["Energy"]

# It is an object data , first convert it into float

data["Energy"] = pd.to_numeric(data["Energy"], errors="coerce")

data = data.dropna(subset=["Energy"])

data.isna().sum()
```

```python
# Finding the outliers in the feature of "Energy"

plt.figure(figsize=(10,3))
sns.boxplot(x=data["Energy"] , data= data )
plt.title("Boxplot of Energy")
plt.show()
```
