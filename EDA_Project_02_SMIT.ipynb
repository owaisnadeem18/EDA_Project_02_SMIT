# **6. How do audio features like danceability, energy, and valence vary across genres?**

```python
data.head()
```

```python
# 1. Convert the 'Danceability', 'Energy', and 'Valence' columns to numeric.
# This fixes the previous 'Could not convert string to numeric' error.
data['Danceability'] = pd.to_numeric(data['Danceability'], errors='coerce')
data['Energy'] = pd.to_numeric(data['Energy'], errors='coerce')
data['Valence'] = pd.to_numeric(data['Valence'], errors='coerce')

# 2. Filter out rows where the 'Genre' is an empty list '[]'.
# This fixes the previous data quality issue.
data_filtered = data[data['Genre'].str.len() > 0]

# 3. Convert the list of genres into a single string.
# This fixes the new 'unhashable type: 'list'' error.
# For example, ['pop', 'rock'] becomes 'pop, rock'.
data_filtered['Genre_string'] = data_filtered['Genre'].str.join(", ")

# 4. Now, perform the groupby and mean calculation using the new 'Genre_string' column.
genre_feature = data_filtered.groupby('Genre_string')[["Danceability", "Energy", "Valence"]].mean()

# Print the resulting DataFrame to see the final, correct output.
print(genre_feature)
```
